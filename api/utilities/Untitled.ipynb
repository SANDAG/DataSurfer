```python
import csv
import json
import urllib2

prod_url = 'http://datasurfer.sandag.org/api'

def getJsonResponse(*args):
    url = '/'.join(args).replace(' ', '%20')
    return json.load(urllib2.urlopen(url))
```

```python
datasource = 'forecast'
series_id = '13'
geography = 'tract'

category = 'income/median'

with open('tract_income_13.csv', 'wb') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    writer.writerow(['tract', 'year', 'median income'])
    for zone_json in sorted(getJsonResponse(prod_url, datasource, series_id, geography)):
        zone = zone_json[geography]
        for data_json in sorted(getJsonResponse(prod_url, datasource, series_id, geography, zone, category)):
            print 'Tract: %s, Year: %s, Median: %s' % (data_json['tract'], data_json['year'], data_json['median_inc'])
            writer.writerow([data_json['tract'], data_json['year'], data_json['median_inc']])
    
```
